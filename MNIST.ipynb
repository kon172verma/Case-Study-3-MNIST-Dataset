```python
# Importing the required libraries.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the MNIST-Training Data into a pandas dataframe.

train_data = pd.read_csv('train.csv')
```

```python
# Printing the top 5 rows of our dataframe.

print(train_data.head(5))
```

```python
# Printing the features and datapoints of our dataset.

print(train_data.shape)
```

```python
# Printing the names of the features and class-label.

print(train_data.columns)
```

```python
# Spliting the dataset into pixel_data and label_data.

label_data = train_data['label']
pixel_data = train_data.drop('label',axis=1)
```

```python
# Printing the rows and columns of our label_data and pixel_data.

print(label_data.shape)
print(pixel_data.shape)
```

```python
# Printing the corresponding image of a datapoint as stated by its features.

plt.figure(figsize=(3,3))
index = 125

grid_data = pixel_data.iloc[index].as_matrix().reshape(28,28)
plt.imshow(grid_data, interpolation = 'none', cmap='gray')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
std_pixel_data = StandardScaler().fit_transform(pixel_data)
print(std_pixel_data.shape)
```

```python
covariance_matrix = np.matmul(std_pixel_data.T, std_pixel_data)
print(covariance_matrix.shape)
```

```python
from scipy.linalg import eigh
eigenvalues, eigenvectors = eigh(covariance_matrix, eigvals=(782,783))
print(eigenvectors.shape)
```

```python
data_2d = np.matmul(std_pixel_data,eigenvectors)
print(data_2d.shape)
```

```python
label_data_2d = np.vstack((label_data,data_2d.T)).T
print(label_data_2d.shape)

label_data_2d = pd.DataFrame(data=label_data_2d, columns=('label','f1','f2'))
print(label_data_2d.head(5))
```

```python
sns.FacetGrid(label_data_2d,hue='label',size=8).map(plt.scatter,'f1','f2').add_legend()
plt.show()
```

```python
from sklearn import decomposition
pca = decomposition.PCA()
```

```python
pca.n_components = 2
pca_data_2d = pca.fit_transform(std_pixel_data)
print(pca_data_2d.shape)
```

```python
label_pca_data_2d = np.vstack((label_data,pca_data_2d.T)).T
print(label_pca_data_2d.shape)

label_pca_data_2d = pd.DataFrame(data=label_pca_data_2d,columns=('label','f1','f2'))
print(label_pca_data_2d.head(5))
```

```python
sns.FacetGrid(label_pca_data_2d,hue='label',size=8).map(plt.scatter,'f1','f2').add_legend()
plt.show()
```

```python
pca.n_components=784
pca_data = pca.fit_transform(std_pixel_data)
```

```python
percentage_variance_explained = pca.explained_variance_ / np.sum(pca.explained_variance_)
cummulative_variance_explained = np.cumsum(percentage_variance_explained)

plt.grid()
plt.xlabel('number_of_components')
plt.ylabel('Cummulative_variance')
plt.plot(cummulative_variance_explained,linewidth=2)
plt.show()
```

```python
from sklearn.manifold import TSNE
```

```python
model = TSNE(n_components = 2, random_state = 0)
```

```python
tsne_data_2d = model.fit_transform(std_pixel_data)
```

```python
label_tsne_data_2d = np.vstack((label_data,tsne_data_2d.T)).T
label_tsne_data_2d = pd.DataFrame(data = label_tsne_data_2d, columns = ('label','Dimension_1','Dimension_2'))
```

```python
sns.FacetGrid(label_tsne_data_2d, hue='label', size=8).map(plt.scatter,'Dimension_1','Dimension_2').add_legend()
plt.show()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=50)
tsne_data_2d = model.fit_transform(std_pixel_data)

label_tsne_data_2d = np.vstack((label_data,tsne_data_2d.T)).T
label_tsne_data_2d = pd.DataFrame(data=label_tsne_data_2d,columns=('label','Dimension_1','Dimension_2'))

sns.FacetGrid(label_tsne_data_2d, hue='label', size=8).map(plt.scatter,'Dimension_1','Dimension_2').add_legend()
plt.show()
```
